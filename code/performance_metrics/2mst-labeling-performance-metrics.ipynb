# Computing Performance Metrics

```python
# read in predictions dataframes
predictions_1 = pd.read_csv("/kaggle/input/my-own-coordinate-data/predictions_1.csv")
predictions_2 = pd.read_csv("/kaggle/input/my-own-coordinate-data/predictions_2.csv")
predictions_3 = pd.read_csv("/kaggle/input/my-own-coordinate-data/predictions_3.csv")
predictions_4 = pd.read_csv("/kaggle/input/my-own-coordinate-data/predictions_4.csv")
```

```python
# read in labels dataframes
labels_1 = pd.read_csv("/kaggle/input/my-own-coordinate-data/labels_1.csv")
labels_2 = pd.read_csv("/kaggle/input/my-own-coordinate-data/labels_2.csv")
labels_3 = pd.read_csv("/kaggle/input/my-own-coordinate-data/labels_3.csv")
labels_4 = pd.read_csv("/kaggle/input/my-own-coordinate-data/labels_4.csv")
```

```python
def compute_metrics_over_datasets(datasets, pred_cols, label_cols):
    
    # initialze empty list for results
    all_results = []
    
    # loop through all dfs
    for i, (df_preds, df_labels) in enumerate(datasets):
        
        # and create dictionary for results
        dataset_results = {'dataset': f'dataset_{i+1}'}
        
        # and compute performance metrics over all tasks
        for pred_col, label_col in zip(pred_cols, label_cols):
            precision = precision_score(df_labels[label_col], df_preds[pred_col])
            recall = recall_score(df_labels[label_col], df_preds[pred_col])
            accuracy = accuracy_score(df_labels[label_col], df_preds[pred_col])
            f1 = f1_score(df_labels[label_col], df_preds[pred_col])
            conf_matrix = confusion_matrix(df_labels[label_col], df_preds[pred_col])
            
            # save results in respective column
            dataset_results[f'precision_{pred_col}'] = precision
            dataset_results[f'recall_{pred_col}'] = recall
            dataset_results[f'accuracy_{pred_col}'] = accuracy
            dataset_results[f'f1_{pred_col}'] = f1
            dataset_results[f'conf_matrix_{pred_col}'] = conf_matrix
        
        # append results dict to all_results
        all_results.append(dataset_results)
    
    # convert into df
    df_results = pd.DataFrame(all_results)
    
    return df_results
```

Combine all individual datasets into one main object:

```python
# store dfs together for convenience
datasets = [
    ('dataset_1', labels_1, predictions_1),
    ('dataset_2', labels_2, predictions_2),
    ('dataset_3', labels_3, predictions_3),
    ('dataset_4', labels_4, predictions_4)
]

# set all different prediction tasks
tasks = ['right_hand', 'left_hand', 'right_knee', 'left_knee', 'right_ground', 'left_ground']

# set colours using Set2 palette for consistency
palette = sns.color_palette("Set2", len(tasks))


plt.figure()

# loop through tasks
for j, task in enumerate(tasks):
    
    # initialize empty arrays for mean FPR and TPR
    mean_fpr = np.linspace(0, 1, 100)
    tpr_values = []
    
    # loop through list of df-tuples and compute ROC
    for dataset_name, labels, predictions in datasets:
        # set label and prediction
        y_true = labels[f'label_{task}']
        y_pred_prob = predictions[f'pred_{task}']
        
        # obtain fpr & tpr
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
        
        # interpolate tpr values using mean fpr and tpr from above
        interp_tpr = np.interp(mean_fpr, fpr, tpr)
        interp_tpr[0] = 0.0
        tpr_values.append(interp_tpr)
    
    # compute mean tpr over datasets
    mean_tpr = np.mean(tpr_values, axis=0)
    mean_tpr[-1] = 1.0
    
    # and compute AUC for mean ROC 
    mean_auc = auc(mean_fpr, mean_tpr)
    
    # have dashed lines for left prediction tasks and solid lines for right prediction tasks
    if 'left' in task:
        linestyle = '--'
    else:
        linestyle = '-'
    
    plt.plot(mean_fpr, mean_tpr, color=palette[j], lw=2, linestyle=linestyle, 
             label=f'Mean {task} (AUC = {mean_auc:.2f})')

# sketch random guess
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# setting and title
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Mean ROC Curves for the Detection of Knee, Hand, and Ground Contact across Datasets')
plt.legend(loc="lower right")

# save figure
plt.savefig('mean_ROC_and_AUC_knee_hand_ground_detection.png')

plt.show()
```
